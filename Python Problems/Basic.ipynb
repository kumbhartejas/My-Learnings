```python
def prime(n):
    count = 0
    num = 2

    while count < n:
        flag = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                flag = False
                break

        if flag:
            print(num, end=" ")
            count += 1

        num += 1

n = int(input("Enter n: "))
prime(n)
```

Check num is armstrong or not (sum of power=len(num) of digits in num == num)

```python
# no== sum of power of digit (power is len)
def armstrong(no):
    str1=str(no)
    power=len(str1)
    total=0
    for i in str1:
      total+=int(i)**power
    if total==no:
        print(f"{no} is armstrong")
    else:
        print(f"{no} is not armstrong")

armstrong(153)
armstrong(143)
```

Find fabonacci series till n

```python
def fibonacci(n):
    fib=[0,1]
    if n<=1:
        return "Number should be greater than 1"
    for i in range(1,n+1):
     fib.append(fib[-2]+fib[-1])
    print(fib)

fibonacci(10)
```

Find count of letters,digits,spaces,symbols in str using re

```python
import re

string="My roll no is 24163."

digits=len(re.findall(r"\d",string))
letters=len(re.findall(r"[A-Za-z]",string))
spaces=len(re.findall(r"\s",string))
symbols=len(re.findall(r"[^A-Za-z0-9\s]",string))
Total=len(re.findall(r"[A-Za-z0-9\s]",string))

print("digits: ",digits)
print("letters: ",letters)
print("spaces: ",spaces)
print("symbols: ",symbols)
print("Total: ",Total)

```

Create lambda func to check types of triange( isoscalas,equilateral,scalen)

```python
x,y,z=10,20,30

isosceles=(lambda a,b,c:a==b or a==c or c==b)
equilateral=(lambda a,b,c: a==b and b==c and c==a)
scalen=(lambda a,b,c: a!=b and b!=c and c!=a)

if isosceles(x,y,z):
    print("It is Isosceles Triangle")
elif equilateral(x,y,z):
    print("It is Equilateral Triangle")
elif scalen(x,y,z):
    print("It is Scalen Triangle")
```

```python
# factorial
def factorial(n):
    if n == 0:      # Base case
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

```python
# Fibonacci

def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    n=fib(n-1) + fib(n-2)
    return n

fib(19)
```
